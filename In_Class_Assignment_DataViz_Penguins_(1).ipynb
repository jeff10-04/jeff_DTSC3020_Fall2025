## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Bill Length vs. Bill Depth by Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.show()
```

**Your interpretation (2–3 sentences):**

The scatter plot shows distinct clusters of bill length and bill depth for each species. Adelie penguins tend to have shorter and deeper bills, while Gentoo penguins have longer and shallower bills. Chinstrap penguins have bill dimensions that fall between Adelie and Gentoo penguins, with a unique shape that helps separate them visually.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=10)
plt.title('Histogram of Bill Length (10 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=30)
plt.title('Histogram of Bill Length (30 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

**Your comparison (1–2 sentences):**

When you increase the number of bins in a histogram, you create more bars, each representing a smaller range of data. This can show more detail in the distribution, revealing peaks and valleys that might be hidden with fewer bins. However, with too many bins, the histogram can become choppy and less smooth, making it harder to see the overall shape of the distribution.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.title('Body Mass by Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```

**Your summary (1–2 sentences):**

Based on the box plot, the Gentoo species has the highest body mass on average, followed by the Chinstrap species, and then the Adelie species with the lowest body mass. The box plot clearly shows the distribution and median body mass for each penguin species, making it easy to compare their weights.
